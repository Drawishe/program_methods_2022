```python
# Импорты

import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
import pydot
from graphviz import Digraph
import os
```

```python
def matrix_change(arr):
    arr[0][1][0]=1
    arr[0][1][1]=3
    arr[0][2][0]=2
    arr[1][2][0]=2
    arr[1][3][0]=1
    arr[1][3][1]=3
    arr[2][0][0]=3
    arr[2][1][0]=1
    arr[2][4][0]=2
    arr[3][3][0]=2
    arr[3][4][0]=1
    arr[3][4][1]=3
    arr[4][2][0]=3
    arr[4][3][0]=2
    arr[4][4][0]=1
    return (arr)

def peak_reverse(peak):
    for i in range(0,5):
        if peak[i] == 1:
            peak[i]=0
        else:
            peak[i]=1
    return(peak)
```

```python
# Создание массива под автомат А'

print('Старт программы')
arr=np.zeros(75, int).reshape(5,5,3)
print('Вывод нулевого массива: ')
print(arr)
```

```python
# Добавление весов и направлений в автомат А', направления указаны от строки к столбцам, 
# q1'=0 элементу массива, q5'=4 элементу массива
# К каждому весу прибавлена +1

matrix_change(arr)

# arr[0][1][0]=1
# arr[0][1][1]=3
# arr[0][2][0]=2
# arr[1][2][0]=2
# arr[1][3][0]=1
# arr[1][3][1]=3
# arr[2][0][0]=3
# arr[2][1][0]=1
# arr[2][4][0]=2
# arr[3][3][0]=2
# arr[3][4][0]=1
# arr[3][4][1]=3
# arr[4][2][0]=3
# arr[4][3][0]=2
# arr[4][4][0]=1

print('Вывод заполненного массива: ')
print(arr)
```
